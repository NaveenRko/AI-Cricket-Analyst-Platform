```python
import pandas as pd
import wikipediaapi
```

```python
players = pd.read_csv("players.csv")
```

```python
players.columns = ['players']
```

```python
players = players["players"].unique().tolist()
len(players)
```

```python
players[:10]
```

```python
wiki = wikipediaapi.Wikipedia(
    user_agent='IPL-RAG',
    language='en'
)
```

```python
# player profiles
count = 0
no_wiki = []
for player in players:

    page = wiki.page(player)

    filename = (
        player
        .lower()
        .replace(" ","_")
    )
    if len(page.text) == 0:
        no_wiki.append(player)
        count = count+1
    with open(
        f"C:/Users/pc/OneDrive/Documents/Data_Science/IPL_multiagent/rag/rag_data/player_profiles/{filename}.txt",
        "w",
        encoding="utf-8"
    ) as f:

        f.write(page.text)
```
